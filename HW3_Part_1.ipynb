***Problem 1:***<br>
Write a program(preferably a function) to scrape the title and price  of all books from this page: http://books.toscrape.com/<br>
Make a DataFrame with the columns "title" and "price" containing the data you scraped.<br>
**Note:** Please, use BeautifulSoup. <br>

```python
###_____________Start code here_________________###
import requests
from bs4 import BeautifulSoup
import pandas as pd


def scrapy_weby(url):
    response=requests.get(url)
    page=response.content
    page=BeautifulSoup(page,"html.parser")
    use=page.find_all("h3")
    titles=[i.find("a").get_text("title-") for i in use]
    prices= page.find_all("p", class_="price_color")
    prices=[i.get_text() for i in prices]
    data=pd.DataFrame({"Titles":titles, "prices":prices})
    return data
        
scrapy_weby("http://books.toscrape.com/")
###_____________End code here_________________###
```
